```python
import pandas as pd

df=pd.read_csv('d:/data/steam/steam2.csv')
df=df.drop(labels=['release_date','platforms','categories','steamspy_tags'], axis=1)
df
```

```python
df2=df[df.columns[0:13]]
df2
```

```python
df_cate=df.loc[:,['Single-player_cate','Steam Achievements_cate','Steam Trading Cards_cate',
'Steam Cloud_cate','Full controller support_cate','Partial Controller Support_cate',
'Multi-player_cate','Steam Leaderboards_cate','Online Multi-Player_cate','Shared/Split Screen_cate']]
df_cate
```

```python
df_steamspy=df.loc[:,['Indie','Action','Casual','Adventure','Strategy','Simulation','Early Access','RPG','Free to Play',
'Puzzle']]
df_steamspy
```

```python
df_y=df[df.columns[-1]]
df_y
```

```python
import pandas as pd

df2=pd.concat([df2,df_cate], axis=1)
df2=pd.concat([df2,df_steamspy], axis=1)
df2=pd.concat([df2,df_y], axis=1)
df2
```
